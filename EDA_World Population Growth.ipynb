## A Question: What continent grows the fastest?

```python
# First we group the population by continent
print('We can already see Asia being the top of the list.')
df.groupby('Continent').sum().sort_values(by='World Population Percentage', ascending = False)
```

```python
# Now we reorder the columns and putting only the population in new dataframe
df2 = df.groupby('Continent')[['1970 Population', '1980 Population', '1990 Population', '2000 Population',
                               '2010 Population', '2015 Population', '2020 Population', '2022 Population',]].sum()
df2
```

```python
# Creating the line plot
plt.figure(figsize=(15, 6))
sbn.lineplot(data=df2.transpose()) # Transpose to change the axis
plt.title("Population Growth by Continent (1970-2022)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend(title="Continent")
plt.show()
```

By using the line plot, we can see the change of population wether it increase or decrease from 1970 to 2022. 
At a glance, we can see that Asia has outstanding growth of population from all the other continent, especially from 1980 to 2010. The road to new milenium shows fast, more than double, population growth to the point Asia's population is far above the others in 2022. The steep increase also helps visualize the outstanding phenomenon.

## The 2022, Population vs Area vs Density

```python
# Styling the plot
plt.figure(figsize=(15, 6))
plt.title("The Population of 2022")
plt.xlabel('Population', fontsize=1)
plt.ylabel('Area', fontsize=1)
sbn.set_style('darkgrid')
sbn.set_context('notebook', font_scale=1)
sbn.despine()

# Filtering 'World Population Percentage' values
filtered_df = df[df['World Population Percentage'] <= 100]  # Only include continents

sbn.scatterplot(data=filtered_df, y='Area (km²)', x='2022 Population', hue='Continent',
                size='World Population Percentage', sizes=(50, 500))

# Finding the furthest dots on each x and y axis
furthest_pointx = df.loc[df['2022 Population'].idxmax()]
furthest_pointy = df.loc[df['Area (km²)'].idxmax()]

# Adding marker and label for the furthest point
plt.scatter(furthest_pointx['2022 Population'], furthest_pointx['Area (km²)'],
            color='black', marker='X', s=10)
plt.annotate(furthest_pointx['Country'], (furthest_pointx['2022 Population'], furthest_pointx['Area (km²)']),
             textcoords="offset points", xytext=(0,10), ha='center')
plt.scatter(furthest_pointy['2022 Population'], furthest_pointy['Area (km²)'],
            color='black', marker='X', s=10)
plt.annotate(furthest_pointy['Country'], (furthest_pointy['2022 Population'], furthest_pointy['Area (km²)']),
             textcoords="offset points", xytext=(0,10), ha='center')

# Customizing the legend for continents
handles, labels = plt.gca().get_legend_handles_labels()
continent_labels = filtered_df['Continent'].unique()

filtered_handles = [h for h, label in zip(handles, labels) if label in continent_labels]
filtered_labels = [label for label in labels if label in continent_labels]

plt.gca().legend(filtered_handles, filtered_labels, loc='upper right', bbox_to_anchor=(1, 1))

plt.show()
```

Looking at the last record of population, year of 2022, by using the scatterplot we can see the population of each country on top of its relationship with the area. The large blue dot on the far right belongs to China and the second one will be India. They are the top 2 of both Asia and the world population itself. But are they the densest country?
